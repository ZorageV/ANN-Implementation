```python
cat_cols = ['artist(s)_name','key', 'mode']

cont_cols = ['artist_count', 'released_year',
       'released_month', 'released_day', 'in_spotify_playlists',
       'in_spotify_charts', 'streams', 'in_apple_playlists', 'in_apple_charts',
       'in_deezer_playlists', 'in_deezer_charts', 'in_shazam_charts', 'bpm','valence_%', 'energy_%',
       'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']

y_cols = ['danceability_%']
```

```python
df = df.dropna()
df = df.reset_index(drop=True)
df.info()
```

```python
cats = np.stack([df[cat].cat.codes.values for cat in cat_cols],1)

cats[:5]
```

```python
df['streams'] = df['streams'].astype('str').astype('int64')
df['in_deezer_playlists'] = df['in_deezer_playlists'].astype('str').str.replace(',', '').astype('int64')
df['in_shazam_charts'] = df['in_shazam_charts'].astype('str').str.replace(',', '').astype('int64')
```

```python
conts = np.stack([df[col].values for col in cont_cols], 1)
conts = torch.tensor(conts, dtype=torch.float)
conts[:5]
```

```python
cat_szs = [len(df[col].cat.categories) for col in cat_cols]
emb_szs = [(size, min(50, (size+1)//2)) for size in cat_szs]
emb_szs
```

```python
selfembeds = nn.ModuleList([nn.Embedding(ni, nf) for ni,nf in emb_szs])
selfembeds
```

```python
embeddingz = []
for i,e in enumerate(selfembeds):
    embeddingz.append(e(cats[:,i]))
embeddingz
```

```python
# We concatenate the embedding sections (12,1,4) into one (17)
z = torch.cat(embeddingz, 1)
z[0].shape
```

```python
class TabularModel(nn.Module):

    def __init__(self,emb_szs,n_cont,out_sz,layers,p=0.3):
        super().__init__()
        self.embds = nn.ModuleList([nn.Embedding(ni,nf) for ni,nf in emb_szs]) #embeddings for categorical data
        self.emb_drop = nn.Dropout(p) #preventing overfitting
        self.bn = nn.BatchNorm1d(n_cont) #scaling

        layerlist = []
        n_emb = sum([nf for ni,nf in emb_szs])
        n_in = n_cont + n_emb

        for i in layers:
            layerlist.append(nn.Linear(n_in,i))
            layerlist.append(nn.ReLU(inplace=True))
            layerlist.append(nn.BatchNorm1d(i))
            layerlist.append(nn.Dropout(p))
            n_in = i
        
        layerlist.append(nn.Linear(layers[-1],out_sz))
        self.layers = nn.Sequential(*layerlist)

    def forward(self,x_cat,x_cont):
        embeddings = []
        for i,e in enumerate(self.embds):
            embeddings.append(e(x_cat[:,i]))
                
        x = torch.cat(embeddings, 1)
        x = self.emb_drop(x)
        x_cont = self.bn(x_cont)
        x = torch.cat([x, x_cont], 1)
        return self.layers(x)
```

```python
model = TabularModel(emb_szs=emb_szs,n_cont=19,out_sz=1,layers=[400,200,100,50,25])
model
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
batch_size = 816
test_size = int(batch_size * .2)

cat_train = cats[:batch_size-test_size]
cat_test = cats[batch_size-test_size:batch_size]
con_train = conts[:batch_size-test_size]
con_test = conts[batch_size-test_size:batch_size]
y_train = y[:batch_size-test_size]
y_test = y[batch_size-test_size:batch_size]
```

```python
len(cat_train)
```

```python
len(cat_test)
```

```python
import time
start_time = time.time()

epochs = 2000
losses = []

for i in range(epochs):
    i+=1
    y_pred = model(cat_train, con_train)
    loss = torch.sqrt(criterion(y_pred, y_train)) # RMSE
    losses.append(loss)
    
    # a neat trick to save screen space:
    if i%25 == 1:
        print(f'epoch: {i:3}  loss: {loss.item():10.8f}')

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    print(f'epoch: {i:3}  loss: {loss.item():10.8f}')

print(f'\nDuration: {time.time() - start_time:.0f} seconds')
```

```python
losses = [loss.item() for loss in losses]
```

```python
plt.plot(range(epochs), losses)
plt.ylabel('RMSE Loss')
plt.xlabel('epoch');
```

```python
with torch.no_grad():
    y_val = model(cat_test, con_test)
    loss = torch.sqrt(criterion(y_val, y_test))
print(f'RMSE: {loss:.8f}')
```
